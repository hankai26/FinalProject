```python
import pandas as pd
```

```python
# Loading Data
file_path = "Comprehensive_Data_List_Uncleaned.csv"
df_static_ds = pd.read_csv(file_path, encoding = 'ISO-8859-1')
df_static_ds.head(10)
```

## Cleaning the DataFrame

```python
# Check datatypes
df_static_ds.dtypes
```

```python
# Drop columns not needed
new_df_static_ds = df_static_ds.drop(['Phone_Number','Fax_Number', 'Email', 'CEO Name', 'CEO Title','CEO Salary'], axis=1)
new_df_static_ds.head()
```

```python
# Check datatypes
new_df_static_ds.dtypes
```

```python
# Turn Zip from float64 to int64 dtype. 
new_df_static_ds['Zip'] = new_df_static_ds['Zip'].astype('Int64')
new_df_static_ds.head()
```

```python
# Check Zip float64 dtype was changed to int64 dtype. 
new_df_static_ds.dtypes
```

```python
# Find null values 
for column in new_df_static_ds.columns:
    print (f"Column {column} has {new_df_static_ds[column].isnull().sum()}null values")
```

```python
# Drop null rows
new_df_static_ds = new_df_static_ds.dropna()
```

```python
# Check there are no null values 
for column in new_df_static_ds.columns:
    print (f"Column {column} has {new_df_static_ds[column].isnull().sum()}null values")
```

```python
# Find duplicate entries
print(f"Duplicate entries: {new_df_static_ds.duplicated().sum()}")
```
