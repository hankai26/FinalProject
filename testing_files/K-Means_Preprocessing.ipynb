```python
# Import Dependencies
import pandas as pd
```

```python
# Data loading
file_path = "Comprehensive_Data_List_Cleaned_AD.csv"
model_data = pd.read_csv(file_path, encoding="ISO-8859-1")
model_data.head(5)
```

```python
# Columns
model_data.columns
```

```python
# List dataframe data types
model_data.dtypes
```

```python
# Find null values 
for column in model_data.columns:
    print (f"Column {column} has {model_data[column].isnull().sum()}null values")
```

```python
# Find duplicate entries
print(f"Duplicate entries: {model_data.duplicated().sum()}")
```

```python
# Drop object dtypes
model_data = model_data.select_dtypes(exclude=['object'])
print(model_data)
```

```python
# Drop 'Zip','Overall Rating' columns
new_model_data = model_data.drop(['Zip','Overall Rating'], axis=1)
new_model_data.head()
```

```python
file_path = "Model_Data_AD.csv"
new_model_data.to_csv(file_path, index=False)
```
