```python
# Cleaning the table
# Create dataframe that only has columns needed and aranging columns
clean_table_df=pd.DataFrame(cn_dataset, columns =['ein', 'charityName','irsClassification__rulingDate','category__categoryName', 'cause__causeName', 'websiteURL', 'mailingAddress__streetAddress1','mailingAddress__streetAddress2','mailingAddress__city','mailingAddress__stateOrProvince', 'mailingAddress__postalCode','currentRating__financialRating__score', 'currentRating__accountabilityRating__score', 'irsClassification__assetAmount', 'irsClassification__incomeAmount''irsClassification__latest990'])
clean_table_df.head()
```

### Convert the data type of rulingDate (from object to datetime)
### Create ruling Year column

```python
clean_table_df["irsClassification__rulingDate"] = pd.to_datetime(clean_table_df["irsClassification__rulingDate"])
clean_table_df.info()
```

```python
clean_table_df["irsClassification__rulingDate"][0:5]
```

```python
ruling_year = clean_table_df["irsClassification__rulingDate"].values.astype('datetime64[Y]').astype(int) + 1970
ruling_year
```

```python
clean_table_df["Ruling_Year"] = ruling_year
clean_table_df.head()
```

```python
# Drop irsClassification__rulingDate column
clean_table_df.drop("irsClassification__rulingDate", axis=1, inplace=True)
clean_table_df.head()
```

```python
# Move 'Ruling_Year' column
cleaner_table_df=pd.DataFrame(clean_table_df, columns =['ein', 'charityName','Ruling_Year','category__categoryName', 'cause__causeName', 'websiteURL', 'mailingAddress__streetAddress1','mailingAddress__streetAddress2','mailingAddress__city','mailingAddress__stateOrProvince', 'mailingAddress__postalCode','currentRating__financialRating__score', 'currentRating__accountabilityRating__score', 'irsClassification__assetAmount', 'irsClassification__incomeAmount''irsClassification__latest990'])
cleaner_table_df.head()
```

```python
cleaner_table_df["Ruling_Year"][0:5]
```

```python
# Adding Present_Year column to subtract from Ruling_Year Colum to get the charity's opperating_years.
cleaner_table_df['Present_Year'] = '2022'
cleaner_table_df.head()
```

```python
# Change the dtype of present year from object into integer
cleaner_table_df["Present_Year"] = cleaner_table_df["Present_Year"].astype(str).astype(int)
```

```python
cleaner_table_df.dtypes
```

```python
# Create new 
cleaner_table_df['Years_In_Opperation'] = cleaner_table_df['Present_Year'] - cleaner_table_df['Ruling_Year']
cleaner_table_df.head()
```

```python
# Create dataframe with only necissary columns for ml model
model_data=pd.DataFrame(cleaner_table_df, columns =['ein','irsClassification__assetAmount','Years_In_Opperation'])
model_data.head()                            
```

```python
# Find null values 
for column in model_data.columns:
    print (f"Column {column} has {model_data[column].isnull().sum()}null values")
```

```python
# Drop null rows
model_data =model_data.dropna()
```

```python
# Check there are no null values 
for column in model_data.columns:
    print (f"Column {column} has {model_data[column].isnull().sum()}null values")
```

```python
# List dataframe data types
model_data.dtypes
```

```python
# Find duplicate entries
print(f"Duplicate entries: {model_data.duplicated().sum()}")
```

```python
model_data.shape
```

```python
# Scale the data - transform irsClassification__assetAmount 
model_data["irsClassification__assetAmount"] = model_data["irsClassification__assetAmount"] / 100000
model_data.irsClassification__assetAmount = model_data.irsClassification__assetAmount.round()
model_data.head()
```

```python
# Saving cleaned data
file_path = "Charity_Navigator_US_States_and_Territories_Cleaned_AD.csv"
model_data.to_csv(file_path, index=False)
```

```python
# Initializing model with K = 4 (since we want four classes of giving levels )
model = KMeans(n_clusters=4, random_state=42)
model
```

```python
# Fitting model
model.fit(model_data)
```

```python
# Get predictions
predictions = model.predict(model_data)
print(predictions)
```

```python
# Add a new class column to the clean_table_df
model_data["class"] = model.labels_
model_data.head()
```

```python
# Create a scatterplot of model_data
model_data.hvplot.scatter(x="irsClassification__assetAmount", y="Years_In_Opperation", by="class")
```

## Elbow Curve to determine best K clusters

```python
inertia = []
k = list(range(1, 11))
```

```python
# Looking for the best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(model_data)
    inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

## Initiate, Fit and Predict Model using Elbow Curve Results
